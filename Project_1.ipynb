**2.2  $min_x||A-XB||^2_F$**

From the definition of Forbenious norm \
$||A-XB||^2_F = \sqrt{Tr((A-XB) * (A-XB)^T)}$ \
Now, \
$Tr((A-XB)*(A-XB)^T) = \\ Tr((A-XB)*(A^T - X^TB^T))=$ \
$Tr(AA^T - AX^TB^T -XBA^T + XBX^TB^T)=$ \
$Tr(AA^T) -Tr(AX^TB^T) -Tr(XBA^T) + Tr(XBX^TB^T)$ \



To minimize this value we should set the derivation of this to 0. \
$ \triangledown_x (Tr(AA^T) -Tr(AX^TB^T) -Tr(XBA^T) + Tr(XBX^TB^T)) = 0  $ \
The derivative of $Tr(AA^T)$ is equal to 0 since it does not include X. \
The derivative of $Tr(AX^TB^T)$ and the derivative of $Tr(XBA^T)$ both are equal to $AB^T$. \
So, we have 
$-2AB^T+ 2XBB^T = 0 $ which is true for $2AB^T$.


# Exercise 3

$f_1([x_1,x_2]^T) = f_1(x_1,x_2) = (x_1-2)^2 + (x_2-3)^2$

```python
import matplotlib.pyplot as plt

def partial_derivative(x,a):
  return (2*(x-a))

values = [] #array used in ploting

curr_x1 = 0 #initial value of x1
curr_x2 = 0
precision = 0.0001
lr = 0.5 #learning rate
T=10 #number of iterations
for i in range(T):
  prev_x1 = curr_x1
  curr_x1 = curr_x1 - lr * partial_derivative(prev_x1, 2 ) #Gradient descent

  prev_x2 = curr_x2
  curr_x2 = curr_x2 - lr *partial_derivative(prev_x2, 3) #Gradient descent

  cost = (curr_x1 -2)**2 + (curr_x2-3)**2
  values.append(cost)
  print("Iteration: ", i ," function value: ", cost,  " x1: ",curr_x1 ," x2: " , curr_x2  )

  diff_x1=abs(curr_x1-prev_x1)
  diff_x2=abs(curr_x2-prev_x2)

  if diff_x1 < precision and diff_x2 < precision :
    print("Minimum: (", curr_x1 ," , ", curr_x2, ") , Reached at iteration ", i+1)
    break;

# Plotting
plt.figure(figsize = (8,6))
plt.plot(values)
plt.xlabel("Iterations")
plt.ylabel("Function Values")

```

**Observations**

For the deafult values on the presentation $x_1,x_2 = (0,0) $ , 
learning rate = 0.5 and T=10 the it needs only two iterations to converge. Results: Minimum: ( 2.0  ,  3.0 ) , Reached at iteration  2

If we try to initialize the variables in a greater values (as we know that the function converge on $x_1 = 2$ and $x_2 =3$) the algorithm also need few iterations to converge. 

For example: 

$x_1,x_2 = (1,2) →$ Minimum: ( 2.0  ,  3.0 ) , Reached at iteration  2 
$x_1,x_2 = (1.5,2.5) →$ Minimum: ( 2.0  ,  3.0 ) , Reached at iteration  2
$x_1,x_2 = (1.9,2.9) →$ Minimum: ( 2.0  ,  3.0 ) , Reached at iteration  2
$x_1,x_2 = (5,6) →$ Minimum: ( 2.0  ,  3.0 ) , Reached at iteration  2

So , for this cost function the initial values of the variables do not play an importrant role.


Learning rate, on the other hand had more significant impact on the algorithm. (To run these examples the number of iteration needs to be encreased, try setting it 10000).
For all this experiments with different learning rates the initial values of variables was $x_1,x_2 = (0,0)$

LR = 0.05 $→$ Minimum: ( 1.999400618654243  ,  2.9991009279813645 ) , Reached at iteration  77

LR = 0.005 $→$ Minimum: ( 1.9934316585421534  ,  2.9901474878132293 ) , Reached at iteration  569

LR = 0.000001 $→$ Fails to converge.

LR = 1 $→$ Fails o converge.

So, we can say that for "extreme" values of learning rate (very small or very big) even a simple cost function like this fails to converge. A s we try to decrease the learning rate the number of iterations needed increases as it is expected.

$f_2([x_1,x_2]^T) = f_2(x_1,x_2) = (1-(x_2-3))^2 +20((x_1+3) - (x_2-3)^2)^2$

*For the default parameters the function fails to converge so the following running corresponds to some other parameters I found to be satisfying for this function. Explained below on "Obseravtions" section.

```python
import matplotlib.pyplot as plt

def partial_derivative(x1,x2,idx): 
  #idx is a parameter to distinguish partial derivative of x1 towards partial derivative of x2
  if idx==1 :
    return (-40*x2**2+240*x2+40*x1-240)
  elif idx==2:
    return (80*x2**3-720*x2**2+1922*x2-80*x2*x1+240*x1-1448)


values = [] #array used in ploting

curr_x1 = -1 #initial value of x1
curr_x2 = 2 #initial value of x2
precision = 0.0001 
lr = 0.0005 #learning rate
T=10000 #number of iterations
for i in range(T):
  prev_x1 = curr_x1
  prev_x2 = curr_x2

  curr_x1 = curr_x1 - lr * partial_derivative(prev_x1,prev_x2,1) #Gradient descent
  curr_x2 = curr_x2 - lr * partial_derivative(prev_x1, prev_x2,2) #Gradient descent

  cost = (1-(curr_x2-3))**2+20*(((curr_x1+3)-(curr_x2-3)**2))**2
  values.append(cost)
  print("Iteration: ", i ," function value: ", cost,  " x1: ",curr_x1 ," x2: " , curr_x2  )

  diff_x1=abs(curr_x1-prev_x1)
  diff_x2=abs(curr_x2-prev_x2)

  if (diff_x1 <= precision and diff_x2 <= precision) :
    print("Minimum: (", curr_x1 ," , ", curr_x2, ") , Reached at iteration ", i+1)
    break;

# Plotting
plt.figure(figsize = (8,6))
plt.plot(values)
plt.xlabel("Iterations")
plt.ylabel("Function Values")
```
